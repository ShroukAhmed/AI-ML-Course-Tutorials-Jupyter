```python
import matplotlib.pyplot as mp
import numpy as np

x= np.array([2,3,5,7,9]) 
y= np.array([4,5,7,10,15])
n= x.size
sum_x=x.sum()
sum_y=y.sum()
sum_xy=sum(x*y) 
mp.scatter(x,y)
mp.plot(x,y, color='red')
mp.show()
m=((n*sum_xy)-(sum_x*sum_y))/((n*sum(x**2))-sum(x)**2)
b=(((sum_y-(m*sum_x))/n))

print (" slope = {m} , vertical axis interception = {b} " .format(m=m, b=b))
```
